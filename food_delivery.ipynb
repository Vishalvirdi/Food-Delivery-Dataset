# Food Orders in New Delhi


## Data loading

### Subtask:
Load the dataset "food_orders_new_delhi (1).csv" into a Pandas DataFrame.


```python
import pandas as pd

df = pd.read_csv("food_orders_new_delhi (1).csv")
display(df.head())
```

## Data cleaning

### Subtask:
Clean the loaded dataset by handling missing values, identifying and addressing outliers, and removing duplicate rows.


```python
import pandas as pd

# Check for missing values
missing_values = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

# Display missing values and their percentages
missing_data = pd.DataFrame({'Missing Values': missing_values, 'Percentage': missing_percentage})
display(missing_data)

# Decide on a strategy for handling missing values
# There are no missing values in this dataset, so no imputation is needed.

# Check for duplicate rows
duplicate_rows_before = df.duplicated().sum()
print(f"Number of duplicate rows before deletion: {duplicate_rows_before}")

# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Check for duplicate rows after deletion
duplicate_rows_after = df.duplicated().sum()
print(f"Number of duplicate rows after deletion: {duplicate_rows_after}")

# Display the info of the cleaned dataframe
print("\nInfo of the cleaned DataFrame:")
df.info()
```

## Data wrangling

### Subtask:
Transform the raw data in the `df` DataFrame into a more usable format by adding new features and adjusting data types.


```python
import pandas as pd

# 1. Create 'Total Price' column
df['Total Price'] = df['Order Value'] + df['Delivery Fee'] - df['Refunds/Chargebacks']

# 2. Convert 'Order Date and Time' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date and Time'])

# 3. Extract 'Day of Week', 'Month', and 'Year'
df['Day of Week'] = df['Order Date'].dt.day_name()
df['Month'] = df['Order Date'].dt.month_name()
df['Year'] = df['Order Date'].dt.year

# 4. Print the info of the df
df.info()
```

## Data exploration

### Subtask:
Explore the dataset to understand its characteristics, including shape, data types, summary statistics, and data distribution.


```python
# Print the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)

# Print the data types of each column
print("\nData types of each column:\n", df.dtypes)

# Generate summary statistics for numerical columns
print("\nSummary statistics for numerical columns:\n", df.describe())

# Generate summary statistics for categorical columns
print("\nSummary statistics for categorical columns:\n", df.describe(include='object'))
```

```python
# Print value counts for 'Payment Method' and 'Discounts and Offers'
print("\nValue counts for Payment Method:\n", df['Payment Method'].value_counts())
print("\nValue counts for Discounts and Offers:\n", df['Discounts and Offers'].value_counts())
```

## Data analysis

### Subtask:
Perform a thorough data analysis to derive insights from the dataset, focusing on order frequency, customer behavior, popular food items, delivery times, and revenue patterns.


```python
import pandas as pd

# Analyze order frequency

# Calculate the number of orders per day
orders_per_day = df.groupby(df['Order Date'].dt.date)['Order ID'].count()
print("Orders per day:\n", orders_per_day)

# Calculate the number of orders per week
orders_per_week = df.groupby(df['Order Date'].dt.isocalendar().week)['Order ID'].count()
print("\nOrders per week:\n", orders_per_week)

# Calculate the number of orders per month
orders_per_month = df.groupby('Month')['Order ID'].count()
print("\nOrders per month:\n", orders_per_month)

# Identify the days, weeks, or months with the highest and lowest order frequencies
print("\nDay with highest order frequency:", orders_per_day.idxmax())
print("Day with lowest order frequency:", orders_per_day.idxmin())
print("\nWeek with highest order frequency:", orders_per_week.idxmax())
print("Week with lowest order frequency:", orders_per_week.idxmin())
print("\nMonth with highest order frequency:", orders_per_month.idxmax())
print("Month with lowest order frequency:", orders_per_month.idxmin())

# Analyze the distribution of orders across different days of the week
orders_by_day_of_week = df['Day of Week'].value_counts()
print("\nDistribution of orders across different days of the week:\n", orders_by_day_of_week)

# Analyze the distribution of orders across different months
orders_by_month = df['Month'].value_counts()
print("\nDistribution of orders across different months:\n", orders_by_month)
```

```python
# Analyze customer behavior

# Calculate the total number of orders placed by each customer
customer_order_counts = df.groupby('Customer ID')['Order ID'].count()
print("Total number of orders placed by each customer:\n", customer_order_counts)

# Determine the average order value per customer
customer_avg_order_value = df.groupby('Customer ID')['Total Price'].mean()
print("\nAverage order value per customer:\n", customer_avg_order_value)

# Identify the top customers based on the number of orders
top_customers_by_orders = customer_order_counts.sort_values(ascending=False).head()
print("\nTop customers based on the number of orders:\n", top_customers_by_orders)

# Identify the top customers based on total spending
customer_total_spending = df.groupby('Customer ID')['Total Price'].sum()
top_customers_by_spending = customer_total_spending.sort_values(ascending=False).head()
print("\nTop customers based on total spending:\n", top_customers_by_spending)

# Analyze the distribution of orders across different customers
print("\nDistribution of orders across different customers:\n", customer_order_counts.describe())

# Analyze the distribution of spending across different customers
print("\nDistribution of spending across different customers:\n", customer_total_spending.describe())
```

```python
# Calculate the number of orders for each restaurant
restaurant_order_counts = df['Restaurant ID'].value_counts()
print("Number of orders for each restaurant:\n", restaurant_order_counts)

# Determine the most popular restaurants
most_popular_restaurants = restaurant_order_counts.head()
print("\nMost popular restaurants:\n", most_popular_restaurants)

# Analyze revenue patterns

# Calculate the total revenue generated per day
revenue_per_day = df.groupby(df['Order Date'].dt.date)['Total Price'].sum()
print("\nTotal revenue per day:\n", revenue_per_day)

# Calculate the total revenue generated per week
revenue_per_week = df.groupby(df['Order Date'].dt.isocalendar().week)['Total Price'].sum()
print("\nTotal revenue per week:\n", revenue_per_week)

# Calculate the total revenue generated per month
revenue_per_month = df.groupby('Month')['Total Price'].sum()
print("\nTotal revenue per month:\n", revenue_per_month)

# Identify the days, weeks, or months with the highest and lowest revenue
print("\nDay with highest revenue:", revenue_per_day.idxmax())
print("Day with lowest revenue:", revenue_per_day.idxmin())
print("\nWeek with highest revenue:", revenue_per_week.idxmax())
print("Week with lowest revenue:", revenue_per_week.idxmin())
print("\nMonth with highest revenue:", revenue_per_month.idxmax())
print("Month with lowest revenue:", revenue_per_month.idxmin())

# Analyze the distribution of revenue across different days of the week
revenue_by_day_of_week = df.groupby('Day of Week')['Total Price'].sum()
print("\nDistribution of revenue across different days of the week:\n", revenue_by_day_of_week)

# Analyze the distribution of revenue across different months
revenue_by_month = df.groupby('Month')['Total Price'].sum()
print("\nDistribution of revenue across different months:\n", revenue_by_month)

# Calculate the average revenue per order
average_revenue_per_order = df['Total Price'].mean()
print("\nAverage revenue per order:", average_revenue_per_order)
```

## Data preparation

### Subtask:
Prepare the data for storytelling by selecting relevant features and aggregating the data as needed.


```python
# 1. Feature Selection
df_story = df[['Order ID', 'Customer ID', 'Restaurant ID', 'Order Date', 'Day of Week', 'Month', 'Total Price', 'Payment Method', 'Discounts and Offers']].copy()

# 2. Data Aggregation: Monthly Revenue
df_monthly_revenue = df_story.groupby('Month')['Total Price'].sum().reset_index()

# 3. Data Aggregation: Weekly Revenue
df_weekly_revenue = df_story.groupby(df_story['Order Date'].dt.isocalendar().week)['Total Price'].sum().reset_index()
df_weekly_revenue.rename(columns={'week': 'Week'}, inplace=True)

# 4. Data Aggregation: Daily Revenue
df_daily_revenue = df_story.groupby(df_story['Order Date'].dt.date)['Total Price'].sum().reset_index()
df_daily_revenue.rename(columns={'Order Date': 'Date'}, inplace=True)

# 5. Data Aggregation: Restaurant Orders
df_restaurant_orders = df_story.groupby('Restaurant ID')['Order ID'].count().reset_index()
df_restaurant_orders.rename(columns={'Order ID': 'Number of Orders'}, inplace=True)

# 6. Data Aggregation: Customer Orders
df_customer_orders = df_story.groupby('Customer ID')['Order ID'].count().reset_index()
df_customer_orders.rename(columns={'Order ID': 'Number of Orders'}, inplace=True)

# 7. Data Aggregation: Customer Spending
df_customer_spending = df_story.groupby('Customer ID')['Total Price'].sum().reset_index()
```
